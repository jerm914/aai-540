```python
import pandas as pd
```

```python
df = pd.read_csv('dataset.csv')
```

```python
print(df.head())
print(df.info())
print(df.describe())
```

```python
# Drop rows with any missing values
df = df.dropna()

# Alternatively, fill missing values with a specific value
# df = df.fillna(value)
```

```python
df = df.drop_duplicates()
```

```python
df
```

```python
# Example: Remove rows where 'energy' is 'invalid_value'
df = df[df['energy'] != 'invalid_value']
```

```python
df
```

```python
# Assuming your DataFrame is named df
df['artists'] = df['artists'].str.replace(',', '')
```

```python
df
```

```python
df.to_csv('cleaned_data.csv', index=False)
```
